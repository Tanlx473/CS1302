```python
def num_prime(lower_bound, upper_bound):
    result = 0
    for i in range(lower_bound, upper_bound):
        flag = 0
        if i == 1 or i == 0:
            continue
        else:
            for j in range(2,i):
                if i % j == 0:
                    flag = 1
                    break
            if flag == 0:
                result += 1
    return result
num_prime(1,10)
```

```python
from math import isqrt

def num_prime(lower_bound, upper_bound):
    def is_prime(n):
        if n<2:
            return False
        for d in range(2, isqrt(n)+1):
            if n%d== 0:
                return False
        else:
            return True
    count = 0
    n = lower_bound
    while n < upper_bound:
        if is_prime(n):
            count +=1
        n+=1
    return count

num_prime(10,20)
```

```python
# Tutorial5 Q6
def traval_distance(hight,c,times):
    if times == None:
        # Summation of Infinite Geometric Series
        return (c+1) * hight / (1-c)
    else:
        # Summation of Geometric Series
        return hight * ((c+1-2*c**(times+1)) /(1-c) - c**times)

hight = 100
c = 0.5
times = 1
print("{:.2f}".format(traval_distance(hight,c,times)))
```

```python
def num_letter_digit(str):
    num_set = set('1234567890')
    letter_set = set('qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM')
    cn = cl = 0
    for i in str:
        if i in num_set:
            cn += 1
        if i in letter_set:
            cl += 1
    return {'DIGITS' : cn, 'LETTERS' : cl}

str1 =  'Hello world! 2020'
num_letter_digit(str1)
```

```python
def shortened(given_list,given_index):
    r = []
    for i in range(len(given_list)):
        if not i in set(given_index):
            r.append(given_list[i])
            
    return r

shortened([12,24,35,70,88,120,155],[0,4,5])
```

```python
def count_character(string):
    rd = {}
    for i in string:
        rd[i] = rd.get(i,0) + 1
    return rd

count_character('abcbabc')
```

```python
def deduped(l):
    return list(set(l))

deduped([1,1,1,2,3,3,4,5,5,6])
```

```python
def sorted_intersection(list1,list2):
    r = []
    for i in list1:
        if i in list2:
            r.append(i)
    rr = sorted(r)
    return rr

print(sorted_intersection([1,2,3],[2,3,4]))
```
